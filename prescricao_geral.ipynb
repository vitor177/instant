```python
# %%
import pandas as pd

df = pd.read_excel('f8c69a2b-71b1-40b0-b70b-0d0b35bfd921.xlsx')
```

```python
df_copy = df.copy()
```

```python
#df.drop_duplicates('%ID_PROCESSO_TRF', inplace=True)

df_copy['Dias restantes para prescrição'] = df_copy['Dias restantes para prescrição'].round(0)

df_copy.info()
```

```python
df_copy['Dias restantes para prescrição']
```

```python
import numpy as np
conditions = [

    (df_copy['Dias restantes para prescrição'] >= 365) & (df_copy['Dias restantes para prescrição'] <= 720),

    (df_copy['Dias restantes para prescrição'] >= 721) & (df_copy['Dias restantes para prescrição'] <= 1095),


    (df_copy['Dias restantes para prescrição'] >= 1096),


    (df_copy['Dias restantes para prescrição'] < 30)

]

values = ['entre_1_e_2', 'entre_2_e_3', 'maior_que_3', '<30']

df_copy['Dias Grupos'] = np.select(conditions, values)

df_copy.head()
```

```python
df_copy.head()
```

```python
tabela_nova = df_copy.groupby(['Localização_Prescrição', 'Dias Grupos'])['Número do processo'].nunique()

tabela_nova
```

```python
tabela_redalert =pd.DataFrame(tabela_nova)

tabela_redalert
```

```python
# 'Localização_Prescrição', 'Dias Grupos'])['Número do processo'
tabela_redalert.reset_index(inplace=True)

tabela_redalert = tabela_redalert.rename(columns={'Localização_Prescrição': 'sigla', 'Dias Grupos':'Alerta', 'Número do processo':'Quantidade'})

#tabela_redalert.set_index(['Seção Judiciária', 'Alerta']).unstack(level=1)['Quantidade']



#tabela_nova.pivot_table('Seção Judiciária', 'Alerta', 'Quantidade', aggfunc='first')


#tabela_redalert.drop(['<30', 'Alerta'], axis=1, inplace=True)

#tabela_redalert.drop('<30', axis=1, inplace=True)

tabela_redalert
```

```python
tabela_redalert = tabela_redalert.pivot_table(index=['sigla'], columns=['Alerta'], values='Quantidade', aggfunc='sum').reset_index()

tabela_redalert
```

```python
tabela_redalert.to_csv('prescricao_resumo.csv', index=False)
```
