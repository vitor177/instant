```python
import pandas as pd

df_copy = pd.read_excel('f8c69a2b-71b1-40b0-b70b-0d0b35bfd921.xlsx')
```

```python
df = df_copy.copy()

estado = 'CE'

# Para testes com quantidade de dias menor que 1 ano
#df.at[df.index[0], 'Dias restantes para prescrição'] = 1
```

```python
df.head()
```

```python
df_ce = df[df['Localização_Prescrição'] == estado]

bins = [0, 365, 730, float('inf')]
labels = ['Menos de 1 ano', 'Entre 1 e 2 anos', 'Mais de 2 anos']
df_ce['Faixa de Tempo para Prescrição'] = pd.cut(df_ce['Dias restantes para prescrição'], bins=bins, labels=labels)

df_ce
```

```python
grouped_df = df_ce.groupby(['Serventia Descrição Prescrição', 'Faixa de Tempo para Prescrição']).size().unstack(fill_value=0)
grouped_df = grouped_df.sort_index()
grouped_df
```
