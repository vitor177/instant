```python
# %%
import pandas as pd

df_copy = pd.read_excel('1df9e396-d7f3-47ec-b3c7-aac4ef1664c8.xlsx')
```

```python
df = df_copy.copy()
```

```python
print(df.columns)

df
```

```python
#df.drop_duplicates('%ID_PROCESSO_TRF', inplace=True)

df['Dias sem movimento'] = df['Dias sem movimento'].round(0)

df.info()
```

```python
df.shape
```

```python
import numpy as np
conditions = [

    (df['Dias sem movimento'] >= 30) & (df['Dias sem movimento'] <= 59),

    (df['Dias sem movimento'] >= 60) & (df['Dias sem movimento'] <= 89),


    (df['Dias sem movimento'] >= 90),


    (df['Dias sem movimento'] < 30)

]

values = ['sem_movimento_30', 'sem_movimento_60', 'sem_movimento_90', '<30']

df['Dias Grupos'] = np.select(conditions, values)

df.head()
```

```python
df['Dias Grupos'].head()

tabela_nova = df.groupby(['Localização', 'Dias Grupos'])['%ID_PROCESSO_TRF'].nunique()

tabela_redalert =pd.DataFrame(tabela_nova)
```

```python


tabela_redalert.reset_index(inplace=True)

tabela_redalert = tabela_redalert.rename(columns={'Localização': 'sigla', 'Dias Grupos':'Alerta', '%ID_PROCESSO_TRF':'Quantidade'})

#tabela_redalert.set_index(['Seção Judiciária', 'Alerta']).unstack(level=1)['Quantidade']

tabela_redalert = tabela_redalert.pivot_table(index=['sigla'], columns=['Alerta'], values='Quantidade', aggfunc='sum').reset_index()


#tabela_nova.pivot_table('Seção Judiciária', 'Alerta', 'Quantidade', aggfunc='first')


#tabela_redalert.drop(['<30', 'Alerta'], axis=1, inplace=True)

tabela_redalert.drop('<30', axis=1, inplace=True)

tabela_redalert
```

```python
#tabela_redalert.drop('Alerta', axis=0, inplace=True)
tabela_redalert
```

```python
tabela_redalert = tabela_redalert.reset_index(drop=True)
#tabela_redalert = tabela_redalert.reindex(['Alerta','Seção Judiciária', 'Processos entre 30 e 59 dias sem movimento', 'Processos entre 60 e 89 dias sem movimento', 'Processos com mais de 90 dias sem movimento'])
tabela_redalert
```

```python
tabela_ordenada = tabela_redalert[['sigla', 'sem_movimento_30', 'sem_movimento_60', 'sem_movimento_90']]
tabela_ordenada

tabela_ordenada
```

```python
colunas_para_formatar = ['Processos entre 30 e 59 dias sem movimento', 'Processos entre 60 e 89 dias sem movimento', 'Processos com mais de 90 dias sem movimento']

for coluna in colunas_para_formatar:
    tabela_ordenada[coluna] = tabela_ordenada[coluna].apply(lambda x: '{:,.0f}'.format(x).replace(',', '.'))

tabela_ordenada
```

```python
datatoexcel = pd.ExcelWriter('testes.xlsx')

tabela_ordenada.to_excel(datatoexcel)

datatoexcel.save()


type(tabela_ordenada)
```
